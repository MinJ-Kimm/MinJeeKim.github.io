# 서울 대피소 현황 지도 만들기                    / Creating a Map of the Seoul Shelter

```python
import pandas as pd
import folium
```

```python
file = '/Users/margokim/Documents/Learning Spoons Data 2/파이썬 기초/Ch3)Data_Visualization/data/서울시 대피소 방재시설 현황 (좌표계_ WGS1984).csv'

raw = pd.read_csv(file, encoding = 'cp949') #utf - 8 
raw
```

```python
raw.head()
```

```python
raw.info()
```

```python


for i in range(len(raw)):

    lat = raw.loc[i , "위도"  ]
    long = raw.loc[i , "경도"]
    name = raw.loc[i, "대피소명칭"]

    print(lat,long,name)
```

```python
m = folium.Map(location = [37.5536067, 126.9674308], zoom_start = 12)

for i in range(len(raw)):

    lat = raw.loc[i , "위도"  ]
    long = raw.loc[i , "경도"]
    name = raw.loc[i, "대피소명칭"]

    folium.Marker([lat,long], tooltip = name).add_to(m)
    
m
```

```python
m.save('/Users/margokim/Documents/Learning Spoons Data 2/파이썬 기초/Ch3)Data_Visualization/data/shelterM.html')
```
